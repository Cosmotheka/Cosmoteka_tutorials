```python
def get_r(cv):
    cv_d = np.diag(cv)
    return cv / np.sqrt(cv_d[:, None] * cv_d[None, :])
```

```python
s = sacc.Sacc.load_fits('../../cataloges/cls_FD_covG.fits')
```
